### Визуализация нормального распределения в зависимости от параметров(loc - среднее в ген. совокупности, scale - среднеквадратическое отклонение в ген. совокупности)

```python
from scipy.stats import norm
```

```python
# выборки из теор. нормального рапределения, соответствующие разным loc и scale
sample_theor_1 = np.linspace(norm.ppf(0.01), norm.ppf(0.99), 100)
sample_theor_2 = np.linspace(norm.ppf(0.01, loc=-1), norm.ppf(0.99, loc=-1), 100)
sample_theor_3 = np.linspace(norm.ppf(0.01, loc=1), norm.ppf(0.99, loc=1), 100)

sample_theor_4 = np.linspace(norm.ppf(0.01), norm.ppf(0.99), 100)
sample_theor_5 = np.linspace(norm.ppf(0.01, scale=2), norm.ppf(0.99, scale=2), 100)
sample_theor_6 = np.linspace(norm.ppf(0.01, scale=3), norm.ppf(0.99, scale=3), 100)
```

```python
fig = plt.figure(figsize=(8, 8))
ax1 = fig.add_subplot(211)
ax1.plot(sample_theor_1, norm.pdf(sample_theor_1), "r", label="loc = 0 scale = 1")
ax1.plot(sample_theor_2, norm.pdf(sample_theor_2, loc=-1), "b", label="loc =- 1 scale = 1")
ax1.plot(sample_theor_3, norm.pdf(sample_theor_3, loc=1), "g", label="lo= + 1 scale = 1")
ax1.grid(True)
ax1.set_title(label="Нормальное распределение для разных параметров среднего в ген. совокупности")
ax1.legend()


ax2 = fig.add_subplot(212)
ax2.plot(sample_theor_4, norm.pdf(sample_theor_4), "r", label="loc = 0 scale = 1")
ax2.plot(sample_theor_5, norm.pdf(sample_theor_5, scale=2), "m", label="loc = 0 scale = 2")
ax2.plot(sample_theor_6, norm.pdf(sample_theor_6, scale=3), "y", label="loc = 0 scale = 3")
ax2.grid(True)
ax2.set_title(label="Нормальное распределение для разных параметров дисперсии в ген. совокупности")
ax2.legend()
```
